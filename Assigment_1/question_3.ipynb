```python
K=[]
expected_rolls_array=[]
for k in range(2,11):
    K.append(k)
    
    probabilities=[1/(2**(k-1))]
    for i in range(2,k):
        probabilities.append(1/(2**(i-1)))
    probabilities.append(1/(2**(k-1)))
    
    expected_rolls=0
    not_seen=k
    for p in probabilities:
        expected_rolls+=1/(p*not_seen)
        not_seen-=1
        
    print(f'k={k}: Expected Rolls={expected_rolls}')
    expected_rolls_array.append(expected_rolls)
```

```python
import matplotlib.pyplot as plt
plt.plot(K, expected_rolls_array, marker='o')
plt.title('Plot')
plt.xlabel('k')
plt.ylabel('Expected Rolls')
plt.grid(True)
plt.show()
```
