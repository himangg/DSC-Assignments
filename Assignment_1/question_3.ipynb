```python
import numpy as np
K=[]
exact_rolls_array=[]
for k in range(2,11):
    K.append(k)
    
    probabilities=[1/(2**(k-1))]
    for i in range(2,k):
        probabilities.append(1/(2**(i-1)))
    probabilities.append(1/(2**(k-1)))
    # print(probabilities)
    found=[]
    found=set(found)
    count=0
    while len(found)<k:
        temp=np.random.choice(range(0, k), p=probabilities)
        # print(temp)
        found.add(temp)
        count+=1
    
    print(f'k={k}: Exact Rolls required={count}')
    exact_rolls_array.append(count)
```

```python
import matplotlib.pyplot as plt
plt.plot(K, exact_rolls_array, marker='o')
plt.title('Plot')
plt.xlabel('k')
plt.ylabel('Exact Rolls')
plt.grid(True)
plt.show()
```
